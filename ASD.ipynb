```python
import pandas as pd
from scipy.io import arff


url = r"C:\aptitude\autistic+spectrum+disorder+screening+data+for+children\Autism-Child-Data.arff"

data, meta = arff.loadarff(url)

df = pd.DataFrame(data)

for col in df.select_dtypes([object]):
    df[col] = df[col].str.decode("utf-8")


df.to_csv("autism_children_dataset.csv", index=False)

print("Dataset saved as autism_children_dataset.csv")
```
